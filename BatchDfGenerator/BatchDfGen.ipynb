```python
import numpy as np
import pandas as pd

df = pd.DataFrame(np.random.rand(17,3), columns=['a', 'b', 'c'])
df
```

```python
from typing import Iterator
class PredFunc:
    @staticmethod
    def df_chunker(df: pd.DataFrame, step_size: int) -> Iterator[tuple[pd.DataFrame, int]]:
        return ((df[pos:pos+step_size], pos) for pos in range(0, df.shape[0], step_size))

step = 4
for idx, (df_batch, pos) in enumerate(PredFunc.df_chunker(df, step)):
    print(f'Batch {idx}: {pos}/{df.shape[0]-1}')
    print(df_batch)
```
